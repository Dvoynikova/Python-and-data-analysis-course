```python
# -    DF -     
Data_2 = Data['Cabin'].value_counts().to_frame()
print('4   : ', Data_2[Data_2['count'] == 4].index.tolist(), '\n',
      '3   : ', Data_2[Data_2['count'] == 3].index.tolist(), '\n',
      '2   : ', Data_2[Data_2['count'] == 2].index.tolist()
)
```

```python
#     
cabin_counts = Data['Cabin'].str.split(' ', expand=True).count(axis=1)
cabin_distribution = cabin_counts.value_counts()
for n, count in cabin_distribution.items():
    print(f"-    -{n} - -  -{count}")
```
