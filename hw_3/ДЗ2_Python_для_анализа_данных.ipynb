количества n-местных кабин (в которых было 2, 3, 4, ... человека)

```python
#считаем кол-во уникальных кабин в DF - столько людей в каждой кабине
Data_2 = Data['Cabin'].value_counts().to_frame()
print('4 человека в кабинах: ', Data_2[Data_2['count'] == 4].index.tolist(), '\n',
      '3 человека в кабинах: ', Data_2[Data_2['count'] == 3].index.tolist(), '\n',
      '2 человека в кабинах: ', Data_2[Data_2['count'] == 2].index.tolist()
)
```

```python
# Подсчет количества мест в кабинках
cabin_counts = Data['Cabin'].str.split(' ', expand=True).count(axis=1)
cabin_distribution = cabin_counts.value_counts()
for n, count in cabin_distribution.items():
    print(f"кол-во мест в кабине -{n} - кол-во человек -{count}")
```
