```python
import pandas as pd
import bottleneck as bn
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/Data/train.csv')
```

```python
%%time
# Рассчитываем средний возраст и стандартное отклонение с помощью bottleneck
average_age = bn.nanmean(data['Age'])
std_age = bn.nanstd(data['Age'])

print(f'Средний возраст: {average_age}')
print(f'Стандартное отклонение возраста: {std_age}')
```

```python
%%time
# Рассчитываем средний возраст и стандартное отклонение с помощью numpy
average_age = np.mean(data['Age'])
std_age = np.std(data['Age'])

print(f'Средний возраст: {average_age}')
print(f'Стандартное отклонение возраста: {std_age}')
```

в данном примере numpy работает быстрее bottleneck

```python
%%time
# Умножаем значение столбца Fare на 1.3 и сохраняем как новый столбец Fare_new
data['Fare_new'] = data['Fare'].apply(lambda x: x * 1.3)

# Проверка результатов
print(data[['Fare', 'Fare_new']].head())
```

```python
%%time
# Умножаем значение Fare на 1.3 для каждого пассажира
for passenger in data.itertuples():
    data.at[passenger.Index, 'Fare_new'] = passenger.Fare * 1.3

# Проверка результатов
print(data[['Fare', 'Fare_new']].head())
```

**Оптимизация типов данных**


Смотрим на минимальное и максимальное значение, и ориентируясь на таблицы выше изменяем тип переменных до минимально допустимого значения

```python
data_housing['price'] = data_housing['price'].astype(np.int32)
data_housing['area'] = data_housing['area'].astype(np.int16)
data_housing['bedrooms'] = data_housing['bedrooms'].astype(np.int8)
data_housing['bathrooms'] = data_housing['bathrooms'].astype(np.int8)
data_housing['stories'] = data_housing['stories'].astype(np.int8)
data_housing['parking'] = data_housing['parking'].astype(np.int8)
```
